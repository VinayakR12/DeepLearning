```python
import matplotlib.pyplot as plt

# 1. Line plot
plt.plot([1,2,3,4], [2,4,6,8])
plt.show()

# 2. Bar chart
plt.bar(["A", "B", "C"], [10, 20, 15])
plt.show()

# 3. Scatter plot
plt.scatter([1,2,3],[3,6,9])
plt.show()

# 4. Histogram
import numpy as np
plt.hist(np.random.randn(100))
plt.show()

# 5. Pie chart
plt.pie([40, 30, 20, 10], labels=["A","B","C","D"])
plt.show()

# 6. Title & labels
plt.plot([1,2,3])
plt.title("My Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()

# 7. Multiple lines
plt.plot([1,2,3], label="Line1")
plt.plot([1,3,5], label="Line2")
plt.legend()
plt.show()

# 8. Box plot
plt.boxplot(np.random.randint(1,100,50))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# 1. Histogram
sns.histplot(df["total_bill"])
plt.show()

# 2. Scatter
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()

# 3. Bar plot
sns.barplot(x="day", y="total_bill", data=df)
plt.show()

# 4. Box plot
sns.boxplot(x="sex", y="tip", data=df)
plt.show()

# 6. Pairplot
sns.pairplot(df)
plt.show()

# 7. Countplot
sns.countplot(x="smoker", data=df)
plt.show()

# 8. Violin plot
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
from scipy import stats
import numpy as np

# 1. Mean
data = [2, 4, 6, 8]
print(stats.tmean(data))

# 2. Mode
print(stats.mode([3, 3, 2, 5]))

# 3. Linear regression
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
print(stats.linregress(x, y))

# 4. Normal distribution PDF
print(stats.norm.pdf(0))

# 5. CDF of normal distribution
print(stats.norm.cdf(1.96))

# 6. Hypothesis test (t-test)
print(stats.ttest_1samp(data, 5))

# 7. Z-score
arr = np.array([10, 12, 15, 18])
print(stats.zscore(arr))

# 8. Integration
from scipy import integrate
result = integrate.quad(lambda x: x ** 2, 0, 2)
print(result)
```
